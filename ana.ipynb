```python
Teams = {  '福爾摩沙台新夢想家':{ 
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    }, 
            '臺北富邦勇士':{
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    }, 
            '新北國王':{
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    }, 
            '桃園領航猿':{
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    }, 
            '高雄鋼鐵人':{
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    }, 
            '新竹街口攻城獅':{
                        'my_total_score':0,
		                'my_total_pos':0,
		                'op_total_score':0,
		                'op_total_pos':0,
		                'AOR':0,
		                'ADR':0,
		                'ANR':0,
		                'Games':[]
                    },
            'PLG':{
                        'total_score':0,
		                'total_pos':0,
		                'AR':0,
                    }
                    
        }
```

```python
def Possessions(FGA, OREB, DDREB, FGM, TOV, FTA):
    return FGA-(OREB/ (OREB+DDREB) ) * (FGA-FGM)*1.07+TOV+0.4*FTA
```

```python
g = [91, 95, 100, 101]
for i in g:
    home_df = pd.read_csv(f"../../Games/{i}/home.csv")
    away_df = pd.read_csv(f"../../Games/{i}/away.csv")

    home_info = home_df.iloc[-1]
    away_info = away_df.iloc[-1]

    home_pos = Possessions((home_info['two']+home_info['trey']), home_info['reb_o'], away_info['reb_d'], (home_info['two_m']+home_info['trey_m']), home_info['turnover'], home_info['ft'])
    away_pos = Possessions((away_info['two']+away_info['trey']), away_info['reb_o'], home_info['reb_d'], (away_info['two_m']+away_info['trey_m']), away_info['turnover'], away_info['ft'])
    home_score = home_info['points']
    away_score = away_info['points']
    home_name = home_info['name_alt']
    away_name = away_info['name_alt']

    # Get the 2&3 points field goals
    home_trey_m = home_info['trey_m']
    home_two_m = home_info['two_m']
    home_trey = home_info['trey']
    home_two = home_info['two']

    away_trey_m = away_info['trey_m']
    away_two_m = away_info['two_m']
    away_trey = away_info['trey']
    away_two = away_info['two']

    # Total
    Teams['PLG']['total_score'] += home_score + away_score
    Teams['PLG']['total_pos'] += home_pos + away_pos

    Teams[home_name]['my_total_score'] += home_score
    Teams[home_name]['my_total_pos'] += home_pos
    Teams[home_name]['op_total_score'] += away_score
    Teams[home_name]['op_total_pos'] += away_pos

    Teams[away_name]['my_total_score'] += away_score
    Teams[away_name]['my_total_pos'] += away_pos
    Teams[away_name]['op_total_score'] += home_score
    Teams[away_name]['op_total_pos'] += home_pos

    # Game
    home_OR = home_score/home_pos * 100
    home_DR = away_score/away_pos * 100
    home_NR = home_OR - home_DR
    home_game = {'my_score': home_score, 'my_pos': home_pos, 'op_score': away_score, 'op_pos': away_pos, 'op_two_m':away_two_m, 'op_two_a':away_two, 'op_trey_m':away_trey_m, 'op_trey_a':away_trey, 'opponent': away_name, 'OR': home_OR, 'DR': home_DR, 'NR': home_NR}
    Teams[home_name]['Games'].append(home_game)

    away_OR = home_DR
    away_DR = home_OR
    away_NR = away_OR - away_DR
    away_game = {'my_score': away_score, 'my_pos': away_pos, 'op_score': home_score, 'op_pos': home_pos, 'op_two_m':home_two_m, 'op_two_a':home_two, 'op_trey_m':home_trey_m, 'op_trey_a':home_trey, 'opponent': home_name, 'OR': away_OR, 'DR': away_DR, 'NR': away_NR}
    Teams[away_name]['Games'].append(away_game)
```

```python
# Calculate average rating for each team and the whole league
for team in Teams:
    if team == 'PLG' and Teams['PLG']['total_pos']!=0:
        Teams['PLG']['AR'] = Teams['PLG']['total_score']/Teams['PLG']['total_pos'] * 100
    elif Teams[team]['my_total_pos']!=0 and Teams[team]['op_total_pos']!=0:
        AOR = Teams[team]['my_total_score']/Teams[team]['my_total_pos']*100
        ADR = Teams[team]['op_total_score']/Teams[team]['op_total_pos']*100
        Teams[team]['AOR'] = AOR
        Teams[team]['ADR'] = ADR
        Teams[team]['ANR'] = AOR - ADR
```
